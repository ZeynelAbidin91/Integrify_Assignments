```python
import numpy as np
import statsmodels.api as sm
import pylab


test = np.random.normal(20, 5, 1000)

sm.qqplot(test, loc = 20, scale = 5, line = '45')
pylab.show()
```

### 1. Why the points don’t lie on the line y=x ?

numpy.random.normal(loc=0.0, scale=1.0, size=None)¶
In "np.random.normal" method, loc specifies the centre of distribution, and scale determines its standart deviation(width). In this example '(20, 0)' is chosen as the centre of distribution while 5 is standart deviation of the distribution. To get points lie on x=y, centre should be 0 and standard deviation should be 1.
A 45-degree reference line is also plotted. If the two sets come from a population with the same distribution, the points should fall approximately along this reference line. The greater the departure from this reference line, the greater the evidence for the conclusion that the two data sets have come from populations with different distributions.

 ### 2.What is the standart of the mean?

    The standard error of the mean, also called the standard deviation of the mean, is a method used to estimate the standard deviation of a sampling distribution. The standard error of the mean refers to the change in mean with different experiments conducted each time.
